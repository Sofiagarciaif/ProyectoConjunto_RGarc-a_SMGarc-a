1. La fórmula el rectángulo se obtiene al interpolar $f(x)$ por medio del polinimio constante $p_0(x)=f(\frac {a+b} {2})$.
<p>&nbsp;</p>
$\int_a^bf(x)dx=(b-a)f(\frac {a+b} {2})=R(f)$
<p>&nbsp;</p>
2. La fórmula del trapecio se obtiene al interpolar a $f$ por medio de un polunomio lineal $p_1(x)=\alpha x + \beta$ que satisfaga $f(a)=p_1(a)$ y $f(b)=p_1(b)$. Determinar el polinimio lineal es equivalente a resolver un sistema de ecuaciones lineales para $\alpha$ y $\beta$ cuya solucion es
<p>&nbsp;</p>
$\alpha=\frac {f(b)-f(a)}{b-a}$ $\beta=\frac {f(b)a-f(a)b}{b-a}$
<p>&nbsp;</p>
Al integrar $p_1(x)$ en el intervalo [a,b] se obtiene la regla del trapecio con $h=b-a$
<p>&nbsp;</p>
$\int_a^bf(x)dx=\int_a^bp_1(x)dx=\frac{h}{2}(f(a)+f(b))=T(f)$
<p>&nbsp;</p>
El error esta dado por 
<p>&nbsp;</p>
$\int_a^bf(x)dx-T(f)=-\frac{h^3}{12}(f^n(n)), a<n<b$
<p>&nbsp;</p>
3. Por ultimo al integrar el polinomio cuadratico que interpola a $f$ en $x=a$, $x=\frac{a+b}{2}$, y $x=b$ se obtiene la regla se Simpson con $h=\frac{b-a}{2}$
<p>&nbsp;</p>
$\int_a^bf(x)dx=\int_a^bp_2(x)dx=\frac{h}{3}(f(a)+4f(a+h)+f(b))=S(f)$
<p>&nbsp;</p>
El error que se comete al usar Simpson es
<p>&nbsp;</p>
$\int_a^bf(x)dx-S(f)=-\frac{h^5}{90}(f^4(n)), a<n<b$
